# MaaSarthi Data Preparation Workflow

This notebook guides the full process of filtering, cleaning, merging, and preparing datasets for India, specifically for MaaSarthi requirements. All steps are performed within the DATASET_NEW folder.

## 1. Import Required Libraries

Import pandas, numpy, and other necessary libraries for data processing.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import os
```

## 2. Load Raw Data Files from DATASET_NEW Folder

Read all relevant raw data files from the DATASET_NEW folder using pandas.

```python
# Example: Load companies.csv and other files
companies = pd.read_csv('archive/companies.csv')
company_industries = pd.read_csv('archive/company_industries.csv')
company_specialities = pd.read_csv('archive/company_specialities.csv')
employee_counts = pd.read_csv('archive/employee_counts.csv')
industries = pd.read_csv('archive/industries.csv')
skills = pd.read_csv('archive/skills.csv')
# Load other files as needed (add more as required)
```

## 3. Filter Data for India

Filter each dataset to include only records related to India.

```python
# Filter companies for India (using country code 'IN')
companies_india = companies[companies['country'].str.upper() == 'IN']

# Filter other datasets by joining with Indian companies
company_ids_india = companies_india['company_id']
company_industries_india = company_industries[company_industries['company_id'].isin(company_ids_india)]
company_specialities_india = company_specialities[company_specialities['company_id'].isin(company_ids_india)]
employee_counts_india = employee_counts[employee_counts['company_id'].isin(company_ids_india)]
# Repeat for other datasets as needed
```
